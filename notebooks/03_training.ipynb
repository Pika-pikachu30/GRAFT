# GRAFT Phase 3: Fine-Tuning Pipeline
Training an open-source model (e.g. Mistral-7B) to reason over graph segments. This relies on QLoRA and TRL.

```python
import sys
import os

sys.path.append(os.path.join("..", "src"))
from graft_trainer import GRAFTTrainer

trainer = GRAFTTrainer(config_path="../config.yaml")
print(f"Base Model: {trainer.config.base_model} | LoRA enabled: {trainer.config.use_lora}")
```

```python
# 1. Load Model with QLoRA BitsAndBytes
model, tokenizer = trainer.setup_model()
print("Model loaded with 4-bit precision")
```

```python
# 2. Setup Dataset and SFTTrainer
# Only run if `../data/processed` dataset exists!
dataset_path = "../data/processed"
if os.path.exists(os.path.join(dataset_path, "train.jsonl")):
    trainer.setup_trainer(dataset_path)
    print("SFTTrainer created successfully!")
else:
    print(f"Dataset not found at {dataset_path}")
```

```python
# 3. Train
# WARNING: This requires significant GPU memory (VRAM). Use an A100 or better for 7B models.
print("Starting training...")
trainer.train()
print("Finished!")
```

```python
# 4. Merge Weights
output_path = "../models/final/"
trainer.export_merged_model(output_path)
print(f"LoRA weights securely merged and saved to {output_path}")
```
