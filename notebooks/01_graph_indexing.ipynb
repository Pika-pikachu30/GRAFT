# GRAFT Phase 1: Hierarchical Knowledge Graph Indexing
This notebook demonstrates how bare text documents are converted into a rich, LLM-extracted Knowledge Graph, clustered into Leiden communities, and hierarchically summarized.

```python
import sys
import os
import networkx as nx

# Add src to path
sys.path.append(os.path.join("..", "src"))
from graph_indexer import GraphIndexer
from data_loader import SyntheticCorpusLoader

indexer = GraphIndexer()

# 1. Load Data
loader = SyntheticCorpusLoader()
texts = loader.preprocess()
print(f"Loaded {len(texts)} documents.")
```

```python
# 2. Chunk Documents
chunks = indexer.chunk_documents(texts[:5], chunk_size=200, overlap=50) # Use small subset for demo
print(f"Produced {len(chunks)} chunks.")
```

```python
# 3. Extract Operations (Entity + Relation)
# Warning: This calls the LLM for every chunk.
graph = indexer.extract_entities_and_relations(chunks[:2], max_glean_iterations=1)
print(f"Graph nodes: {graph.number_of_nodes()}, edges: {graph.number_of_edges()}")
```

```python
# 4. Community Detection
hierarchy = indexer.detect_communities(graph)
print(f"Hierarchy levels: {list(hierarchy.keys())}")
```

```python
# 5. Summarize Communities
summaries = indexer.summarize_communities(hierarchy, graph)
for cid, summary in summaries.items():
    print(f"--- {summary.title} (Level {summary.level}) ---")
    print(summary.summary[:150] + "...\n")
```

```python
# 6. Save
indexer.save_index("../data/graph_index")
print("Index Saved!")
```
